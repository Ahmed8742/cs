```python
import turtle as tr
import random as rd

curser = tr.Turtle()
curser.speed(100)

for i in range(300):
    curser.circle(15 + i, 100)
    first_color = rd.random()
    second_color = rd.random()
    third_color = rd.random()
    curser.pencolor({first_color, second_color, third_color})

curser.hideturtle()
tr.done()
```

```python
import turtle as tr

curser = tr.Turtle()
curser.speed(100)

curser.penup()
curser.right(90)
curser.forward(200)
curser.right(90)
curser.forward(200)
curser.right(180)
curser.pendown()

i = 0
for x in range(8):
    while i < 72:
        curser.begin_fill()
        curser.forward(50)
        curser.left(90)
        curser.forward(50)
        curser.left(90)
        curser.forward(50)
        curser.left(90)
        curser.forward(50)
        curser.left(90)
        curser.penup()
        curser.forward(50)
        curser.pendown()
        if i % 2 == 0:
            curser.end_fill()
        i += 1
        if i in [8, 17, 26, 35, 44, 53, 62, 71]:
            i += 1
            break
    curser.right(180)
    curser.forward(400)
    curser.right(90)
    curser.forward(50)
    curser.right(90)
curser.hideturtle()
    
    
tr.done()
```
